```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values from -5 to 5
x = np.linspace(-5, 5, 1000001)

# Initialize an empty array to store the function values
y = []

# Calculate the function values based on the conditions
for i in x:
    if i < 0:
        y.append(i * np.sin(i**2))
    else:
        y.append(np.cos(i))

# Plot the function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the Function f(x)')
plt.grid(True)
plt.show()
```
